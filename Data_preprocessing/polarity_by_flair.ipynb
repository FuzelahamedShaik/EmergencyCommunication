```python
from flair.models import TextClassifier
from flair.data import Sentence
from segtok.segmenter import split_single
import pandas as pd
import re
```

```python
classifier = TextClassifier.load('en-sentiment')
```

```python
def make_sentences(text):
    """ Break apart text into a list of sentences """
    sentences = [sent for sent in split_single(text)]
    return sentences
```

```python
def predict(sentence):
    """ Predict the sentiment of a sentence """
    if sentence == "":
        return 0
    text = Sentence(sentence)
    # stacked_embeddings.embed(text)
    classifier.predict(text)
    value = text.labels[0].to_dict()['value'] 
    if value == 'POSITIVE':
        result = text.to_dict()['all labels'][0]['confidence']
    else:
        result = -(text.to_dict()['all labels'][0]['confidence'])
    return round(result, 3)
```

```python
def get_scores(sentences):
    """ Call predict on every sentence of a text """
    results = []
    for i in range(0, len(sentences)): 
        results.append(predict(sentences[i]))
    return results

def get_sum(scores):
    
    result = round(sum(scores), 3)
    return result
```

```python
df = pd.read_csv('C:\\Users\\fuzel\\Documents\\Master Thesis\\Data_extraction\\Dump\\Emergency plus play store.csv')
df = df.dropna()
df['sentences'] = df['Review'].apply(make_sentences)
```

```python
df['scores'] = df['sentences'].apply(get_scores)
df['scores_sum'] = df.scores.apply(get_sum)
```

```python
senti_list = []
for value in df['scores_sum']:
    if value < 0:
        senti_list.append(-1)
    else:
        senti_list.append(1)
df['sentiment'] = senti_list
```
